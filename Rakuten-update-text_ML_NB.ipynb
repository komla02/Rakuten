### Nettoyage des Données Textuelles


```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
# Téléchargement des ressources NLTK 
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
# Chargement des données
X = pd.read_csv("C:\\Users\\user\\OneDrive\\DATASIENCETEST\\PROJET\\Data\\Update\\X_train_update.csv", index_col=0)
y = pd.read_csv("C:\\Users\\user\\OneDrive\\DATASIENCETEST\\PROJET\\Data\\Update\\Y_train_CVw08PX.csv", index_col=0).squeeze().map(str)
```

```python
# Fonction de nettoyage de texte
def nettoyer_texte(texte):
    # Vérification si le texte est non vide et ne ressemble pas à un chemin de fichier
    if texte and not re.match(r'\s*[/\\]', texte):
        try:
            # Suppression des balises HTML
            texte = BeautifulSoup(texte, "html.parser").get_text()
        except MarkupResemblesLocatorWarning:
            pass  # Si l'avertissement est levé, conservez le texte tel quel
    
    # Suppression des caractères non alphabétiques et normalisation
    texte = re.sub(r'[^a-zA-Z]', ' ', texte)
    mots = word_tokenize(texte.lower())
    
    # Lemmatisation et suppression des stop words
    stop_words_francais = set(stopwords.words('french'))
    mots_finaux = [WordNetLemmatizer().lemmatize(mot) for mot in mots if mot not in stop_words_francais]
    
    return ' '.join(mots_finaux)
```

```python
# Fusion des colonnes 'designation' et 'description' avec gestion des NaN
X['text_combined'] = X['designation'].fillna('') + ' ' + X['description'].fillna('')
X['text_clean'] = X['text_combined'].apply(nettoyer_texte)
```

```python
# Définition du chemin de sauvegarde basé sur le chemin des fichiers originaux
chemin_sauvegarde = "C:\\Users\\user\\OneDrive\\DATASIENCETEST\\PROJET\\Data\\Update\\X_clean_before_modeling.csv"

# Sauvegarde du DataFrame nettoyé
X.to_csv(chemin_sauvegarde, index=False)
```

```python
# Vérification de l'existence des colonnes avant de les supprimer pour éviter des erreurs
colonnes_a_supprimer = ['designation', 'description', 'text_combined']
for colonne in colonnes_a_supprimer:
    if colonne in X.columns:
        X.drop(colonne, axis=1, inplace=True)
```

```python
display(X)
```

```python
# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X['text_clean'], y, test_size=0.2, random_state=42)
```

```python
# Initialisation et application du vectoriseur TF-IDF
vectorizer = TfidfVectorizer(max_features=10000, stop_words=stopwords.words('french'))
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)
```

```python
# Entraînement du modèle Multinomial Naive Bayes
model = MultinomialNB()
model.fit(X_train_vect, y_train)

```

```python
# Prédiction et évaluation
predictions = model.predict(X_test_vect)
print(classification_report(y_test, predictions))
```
